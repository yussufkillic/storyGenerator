```python
#gpt2 modelini başlatıyor ve hugging face'den import ediyoruz.

from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

model
```

```python
#modeli test ediyoruz

prompt = "Once upon a time"

inputs = tokenizer(prompt, return_tensors="pt")
outputs = model.generate(inputs.input_ids, max_new_tokens=100, do_sample=True, top_k=50, top_p=0.95)

outputs
```

```python
# veri setini hugging face'den import ediyoruz

from datasets import load_dataset

short_stories_dataset = load_dataset("roneneldan/TinyStories")
short_stories_dataset
```

```python
short_stories_dataset["train"][2]
```

```python
# veri kümemizden daha küçük bir veri seti kullanıyoruz

small_story_dataset = load_dataset("roneneldan/TinyStories", split="train[:1000]")
small_story_dataset = small_story_dataset.train_test_split(train_size=0.8)

small_story_dataset
```

```python
[len(x["text"].split(" ")) for x in small_story_dataset["train"]]
```

```python
# veri setimizi tokenize ediyoruz

def preprocess_batch(batch):
  all_text_items = batch["text"]
  trimmed_text_items = [x[:500] for x in all_text_items]
  return tokenizer(trimmed_text_items)

tokenized_dataset = small_story_dataset.map(
    preprocess_batch,
    batched=True,
    batch_size=10,
    remove_columns=small_story_dataset["train"].column_names,
)

tokenized_dataset
```

```python
tokenized_dataset["train"][0]
```

```python
#data collator oluşturuyoruz
from transformers import DataCollatorForLanguageModeling

tokenizer.pad_token = tokenizer.eos_token
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
data_collator
```

```python
# Eğitim için gerekli kütüphaneleri içe aktaralım
from transformers import TrainingArguments, Trainer

# Eğitim argümanlarını oluşturalım
training_args = TrainingArguments(
    output_dir="./output",  # Çıkış dizini
    eval_strategy="epoch",  # evaluation_strategy yerine eval_strategy
    learning_rate=2e-5,     # Öğrenme oranı
    weight_decay=0.01,      # Ağırlık çürümesi
    num_train_epochs=1,    # Eğitim epoch sayısı
    logging_dir="./logs",   # Log dosyalarının kaydedileceği dizin
    run_name="my_experiment",  # W&B için özel bir çalışma ismi
)

# Trainer sınıfını tanımlayalım
trainer = Trainer(
    model=model,  # Modelinizi buraya ekleyin
    train_dataset=tokenized_dataset["train"],  # Eğitim verisi
    eval_dataset=tokenized_dataset["test"],    # Test verisi
    args=training_args,  # Eğitim argümanları
    data_collator=data_collator  # Veri düzenleyici
)

# Weights & Biases (W&B) devre dışı bırakma (isteğe bağlı)
import os
os.environ["WANDB_DISABLED"] = "true"

# Eğitimi başlatalım
trainer.train()
```
